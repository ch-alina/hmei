```python
import numpy as np
from matplotlib import pyplot as plt

data = np.loadtxt('/home/ac63/storage/PF.csv', delimiter=',', skiprows=1)
lon_pf = data[:,0]
lat_pf = data[:,1]

data2 = np.loadtxt('/home/ac63/storage/SAF.csv', delimiter=',', skiprows=1)
lon_saf = data2[:,0]
lat_saf = data2[:,1]

print(max(lat_saf))

import cartopy.crs as ccrs
import cartopy

fig = plt.figure(figsize=(12, 12))
projection = ccrs.PlateCarree(central_longitude=0)
ax = plt.axes(projection=projection)
ax.coastlines()
ax.set_extent((-180, 180, -90, 60), crs=ccrs.PlateCarree())
ax.scatter(lon_pf, lat_pf, s=2, color='red')
ax.scatter(lon_saf, lat_saf, s=2, color='black')
```
